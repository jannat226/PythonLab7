```python
# ques1  Create two 3×3 matrices using the random function in Numpy and perform the following operations.
# è Product (prod)
# è Multiplication (multiply)
# è Dot Product (dot)
import numpy as np
from numpy import random

x = random.rand(3, 3)
y = random.rand(3, 3)
print("Matrix x is",x)
print("Matrix y is",y)
res=np.multiply(x,y)
print("The multiplication of matrix using multiply is ",res)
res=np.prod([x,y])
print(res)
res=np.dot(x,y)
print("The product of matrix using dot is",res)


# ques2 Perform the following set operations using the Numpy functions.
# è Union
# è Intersection
# è Set difference
# è XOR

arr1 = np.array([1, 2, 3, 4])
arr2 = np.array([3, 4, 5, 6])

res = np.union1d(arr1, arr2)
print("\n\n\nThe union of two arr1 and arr2",res)

res = np.intersect1d(arr1, arr2, assume_unique=True)
print("\n\n\nThe intersection of two arr1 and arr2",res)

res = np.setdiff1d(arr1, arr2, assume_unique=True)
print("\n\n\nThe set difference of two arr1 and arr2",res)

res = np.setxor1d(arr1,arr2, assume_unique=True)
print("The set difference of two arr1 and arr2",res)
# ques3Q3. Create a 1D array using Random function and perform the following operations.
# è Cumulative sum
# è Cumulative Product
# è Discrete difference (with n=3)
# è Find the unique elements from the array
res = np.cumsum([arr1,arr2])
print("\n\n\nCumulative sum",res)
res = np.cumprod([arr1,arr2])
print("\n\n\nCumulative product",res)
res = np.diff(arr1)
print("\n\n\nCumulative difference",res)

res = np.unique(arr1)
print("\n\n\nUnique element of the array",res)

# ques4 Create two 1D array and perform the Addition using zip(), add() and user defined function (frompyfunc())


x = [1, 2, 3, 4]
y = [4, 5, 6, 7]
z = []

for i, j in zip(x, y):
  z.append(i + j)
print("zip od array x and y ",z)

x = [1, 2, 3, 4]
y = [5,6,7,8]
res = np.add(x, y)
print("addition of element of the 2 arrays",res)


def myadd(x, y):
  return x+y

myadd = np.frompyfunc(myadd, 2, 1)

print("user defined function ",myadd([1, 2, 3, 4], [5, 6, 7, 8]))

# Q5. Find the LCM (Least Common Multiple) and GCD (Greatest Common Divisor) of an array of elements using reduce().
arr = np.array([3, 6, 9])
x = np.lcm.reduce(arr)
print("LCM of num1 and num2",x)


arr = np.array([3, 6, 9])
x = np.gcd.reduce(arr)
print("GCD of the elements of the array",x)
```
